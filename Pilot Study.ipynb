```python
dfs = []
for fn in files:
    try:
        df = pd.read_csv(os.path.join('all_libguides', fn))
        df['Site'] = fn
        dfs.append(df)
    except Exception as e:
        print('Error loading: ', fn)
        print(e)
```

```python
#df
```

```python
len(dfs)
```

```python
df = pd.concat(dfs)
```

```python
df.columns
```

```python
df[:5]
```

```python
len(df)
```

```python
df = df[['Category', 'Link', 'Subject', 'Site']]
```

```python
df = df.drop_duplicates()
```

```python
len(df)
```

```python
subject_counter = df['Subject'].value_counts()
```

```python
subject_counter
```

```python
df_hkbu = pd.read_csv('8 universities from HK/HKBU.libguides.csv')
```

```python
df_hkbu['Subject']
```

```python
hkbu_subject_set = set(df_hkbu['Subject'].apply(lambda x: x.replace('Subject Guide', '')).values)
```

```python
hkbu_subject_set
```

```python
hkbu_missing_guides = set(subject_counter[:50].keys()) - hkbu_subject_set
hkbu_missing_guides
```

```python
hkbu_refs = df[df['Subject'].apply(lambda x: x in hkbu_missing_guides)][
    ['Link', 'Subject']
].drop_duplicates().sort_values(by='Subject').set_index('Subject')
hkbu_refs.to_csv('hkbu-reference.csv')
hkbu_refs
```
